  **Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
coro_set = pd.read_csv("covid19_Confirmed_dataset.csv")
coro_set.head()
```

```python
coro_set.shape
```

```python
coroSinLat = coro_set.drop(["Lat", "Long"], axis=1)
```

```python
coroSinLat.head(10)
```

```python
coroAgregatted = coroSinLat.groupby("Country/Region").sum()
```

```python
coroAgregatted.head(10)
```

```python
coroAgregatted.shape
```

```python
coroAgregatted.loc["France"].plot()
coroAgregatted.loc["Colombia"].plot()
coroAgregatted.loc["China"].plot()
plt.legend()
```

**Find max infection rate for all countries

```python
countries = list(coroAgregatted.index)
maxPerCountry = []
for c in countries:
    maxPerCountry.append(coroAgregatted.loc[c].diff().max())
coroAgregatted["Max infection rate"] = maxPerCountry
```

```python
coroAgregatted.head(10)
```

**Show only max Infection rate

```python
coro_data = pd.DataFrame(coroAgregatted["Max infection rate"])
```
